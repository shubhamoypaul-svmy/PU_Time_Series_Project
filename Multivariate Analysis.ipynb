```python
## split data into 80-20 percentage
from sklearn.model_selection import train_test_split
x = df.iloc[:,[10,11,13]]
y = df.iloc[:,9]
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.2,shuffle=False)
xtrain.shape, xtest.shape, ytrain.shape, ytest.shape
```

# Simple Linear Regression

```python
from sklearn.linear_model import LinearRegression
# Create a linear regression model and fit the data
model = LinearRegression().fit(xtrain, ytrain)

print(model.intercept_, model.coef_)

y_pred = model.predict(xtest)
```

```python
import matplotlib.pyplot as plt

index = range(len(ytest))
# Actual values
actual = ytest

# Predicted values
predicted = y_pred

# Create a figure and axis object
plt.plot(index,actual, label = "actual")
plt.plot(index,predicted, label = "predicted")


# Set plot title and labels
plt.title('Actual vs. Predicted Values')
plt.xlabel('Index')
plt.ylabel('Value')

# Add legend
plt.legend()
plt.gca().invert_xaxis()
plt.savefig("figures/linear_reg.svg")

# Display plot
plt.show()
```

```python
from sklearn.metrics import r2_score

# Calculate R-squared predicted
r2_pred = r2_score(actual, predicted)

print("R-squared:", round(r2_pred,2))
```

```python
from sklearn.metrics import mean_squared_error
import math


# calculate mean squared error
mse = mean_squared_error(actual, predicted)

# calculate root mean squared error
rmse = math.sqrt(mse)
print("MSE:", mse)
print("RMSE:", rmse)
```

```python
# calculate aic
def calculate_aic(n, mse, num_params):
 aic = n * log(mse) + 2 * num_params
 return aic

def calculate_bic(n, mse, num_params):
 bic = n * log(mse) + num_params * log(n)
 return bic
```

```python
from sklearn.metrics import mean_squared_error
from math import log
import math
from sklearn.metrics import r2_score

# calculate the error
mse = mean_squared_error(actual, predicted)
print('MSE: %.3f' % mse)
# calculate the aic
aic = calculate_aic(len(actual), mse,3)
print('AIC: %.3f' % aic)
# calculate the aic
bic = calculate_bic(len(actual), mse,3)
print('BIC: %.3f' % bic)
# Return the square root 
print ('RMSE: %.3f' % math.sqrt(mse))
```

```python
print(huber.intercept_, huber.coef_)
```

```python
import matplotlib.pyplot as plt

index = range(len(ytest))
# Actual values
actual = ytest

# Predicted values
predicted = huber.predict(xtest)

# Create a figure and axis object
plt.plot(index,actual, label = "actual")
plt.plot(index,predicted, label = "predicted")


# Set plot title and labels
plt.title('Actual vs. Predicted Values')
plt.xlabel('Index')
plt.ylabel('Value')

# Add legend
plt.legend()
plt.gca().invert_xaxis()
# save plot
plt.savefig("figures/huber_reg.svg")

# Display plot
plt.show()
```
